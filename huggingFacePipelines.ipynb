```python
# Classification

classifier = pipeline("zero-shot-classification", device="cuda")
result = classifier("Hugging Face's Transformers library is amazing!", candidate_labels=["technology", "sports", "politics"])
print(result)
```

```python
from transformers import pipeline

# Load the classifier
classifier = pipeline("zero-shot-classification", device="cuda")

# Technology Example
tech_text = "Artificial Intelligence is revolutionizing industries worldwide, enabling automation, data-driven decisions, and innovative solutions that were unimaginable a decade ago."
tech_result = classifier(tech_text, candidate_labels=["technology", "sports", "politics"])
print(tech_result)

# Dog Speech Example
dog_text = "Dogs are loyal companions that bring joy and happiness to their owners. They communicate through barking, body language, and even facial expressions, making them deeply connected to human emotions."
dog_result = classifier(dog_text, candidate_labels=["animals", "psychology", "science"])
print(dog_result)

# Sentiment Analysis Example
sentiment_text = "The customer service was absolutely terrible. I had to wait for over an hour, and when I finally got help, the representative was rude and unhelpful."
sentiment_result = classifier(sentiment_text, candidate_labels=["positive", "negative", "neutral"])
print(sentiment_result)
```
